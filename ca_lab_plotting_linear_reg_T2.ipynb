```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import minmax_scale
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import QuantileTransformer
import statsmodels.api as sm
from scipy.stats import boxcox
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.linear_model import Lasso,Ridge,ElasticNet, LinearRegression

pd.options.display.max_rows = 50
```

```python
df_comp = pd.read_csv('Data_Marketing_Customer_Analysis_Round2.csv')
```

```python
## Effective to date column
pd.to_datetime(df_comp["Effective To Date"],format="%m/%d/%y")
df_comp['Effective To Date'] = pd.DatetimeIndex(df_comp['Effective To Date']).month
```

```python
df_comp
```

```python
#convert ETD column to object type
df_comp['Effective To Date'] = df_comp['Effective To Date'].astype('object')
df_comp['Effective To Date']
```

```python
# drop uninformative columns
df_comp.drop(columns=["Unnamed: 0", "Customer", "Vehicle Type"], inplace = True)
```

```python
# numerical variables found informative in T1 Lin Reg model: boxplot of MPA
sns.boxplot(df_comp['Monthly Premium Auto'])
```

```python
# MPA: removing outliers
Q1 = np.percentile(df_comp['Monthly Premium Auto'], 25)
Q3 = np.percentile(df_comp['Monthly Premium Auto'], 75)
IQR = Q3 - Q1

df_clean = df_comp[(df_comp['Monthly Premium Auto']  < (Q3+1.5*IQR))]
df_clean
sns.boxplot(df_clean['Monthly Premium Auto'])
```

```python
# Resetting row index
df_clean = df_clean.reset_index(drop=True)
```

```python
df_clean
```

```python
# Replace 0 with nan: Income column
df_clean['Income'] = df_clean['Income'].replace(0,np.nan)
sns.histplot(df_clean['Income'])
```

```python
# Removing NaNs!
df_clean = df_clean.dropna(axis = 0, how = 'any')
df_clean = df_clean.reset_index(drop=True)
df_clean
```

```python
# Splitting columns to categorical vs. numerical
numeric_int = df_clean.select_dtypes(include = int)
numeric_int
numeric_float = df_clean.select_dtypes(include = float)
numeric_float
df_numerical = pd.concat([numeric_int,numeric_float], axis=1)
df_numerical
```

```python
plt.figure(figsize=(20,12))
for i, column in enumerate(df_numerical.columns,1):
    plt.subplot(3,3,i)
    sns.histplot(df_numerical[column])
```

```python
# Categorical
df_cat = df_clean.select_dtypes(include = object)
df_cat
```

```python
# Get dummy variables: categorical variables
categorical_dummy = pd.get_dummies(df_cat[:], drop_first=True)
categorical_dummy
```

```python
## Linear regression T2: prep
```

```python
# Concatenating numerical columns with dummy categorical columns: indef. vars.

X = df_numerical.iloc[:,:-1]
X = pd.concat([X,categorical_dummy], axis=1)
X
```

```python
# Dep. var.
y = df_numerical.iloc[:,-1]
y
```

```python
# Train/Test split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=.30,random_state=100)
```

```python
# Fitting train sample and scaling both samples
std_scaler = StandardScaler().fit(X_train) 
X_train_scaled = std_scaler.transform(X_train)
X_test_scaled=std_scaler.transform(X_test)

print(X_train_scaled)
print("--------")
print(X_test_scaled)
```

```python
# of Income X_train applying Quantile Transformer
qt = QuantileTransformer(output_distribution='normal')
qt.fit(X_train['Income'].to_numpy().reshape(-1,1))
X_train['Income']= qt.transform(X_train['Income'].to_numpy().reshape(-1,1))
sns.displot(X_train['Income'])
```

```python
# of Income test applying Quantile Transformer
X_test['Income'] = qt.transform(X_test['Income'].to_numpy().reshape(-1,1))
sns.displot(X_test['Income'])
```

```python
##of Customer Lifetime Value applying Quantile Tranformer 
qt = QuantileTransformer(output_distribution='normal')
qt.fit(X_train['Customer Lifetime Value'].to_numpy().reshape(-1,1))
X_train['Customer Lifetime Value']= qt.transform(X_train['Customer Lifetime Value'].to_numpy().reshape(-1,1))
sns.displot(X_train['Customer Lifetime Value'])
```

```python
#of Monthly Premium Auto X_train with Yeo-Johnson
yj = PowerTransformer(method="yeo-johnson")
yj.fit_transform(X_train['Monthly Premium Auto'].to_numpy().reshape(-1,1))
X_train['Monthly Premium Auto']= yj.fit_transform(X_train['Monthly Premium Auto'].to_numpy().reshape(-1,1))
sns.displot(X_train['Monthly Premium Auto'])
```

```python
# of Monthly Premium Auto X_test with Yeo-Johnson
yj.fit_transform(X_test['Monthly Premium Auto'].to_numpy().reshape(-1,1))
X_test['Monthly Premium Auto']= yj.fit_transform(X_test['Monthly Premium Auto'].to_numpy().reshape(-1,1))
sns.displot(X_test['Monthly Premium Auto'])
```

# Comparing regression models 

```python
def model_comparison(a, b, c, d):
    models = [Lasso(), Ridge(), LinearRegression()]
    for model in models:
        model.fit(a, c)
        print(f"{model.__class__.__name__}: Train -> {model.score(a, c)}, Test -> {model.score(b, d)}") 
```

```python
model_comparison(X_train_scaled, X_test_scaled, y_train, y_test)
```
